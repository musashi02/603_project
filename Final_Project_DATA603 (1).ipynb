Create a spark data frame

```python
df_spark = spark.createDataFrame(df)
```

```python
from pyspark.sql.functions import col, to_timestamp, hour, dayofweek, year

# Drop unnecessary columns
drop_cols = ["Unnamed: 0", "first", "last", "street", "trans_num"]
df_spark = df_spark.drop(*drop_cols)

# Convert 'trans_date_trans_time' and 'dob' to timestamp
df_spark = df_spark.withColumn("trans_date_trans_time", to_timestamp("trans_date_trans_time"))
df_spark = df_spark.withColumn("dob", to_timestamp("dob"))

# Extract hour and day of week
df_spark = df_spark.withColumn("trans_hour", hour("trans_date_trans_time"))
df_spark = df_spark.withColumn("trans_dayofweek", dayofweek("trans_date_trans_time"))

# Calculate age
df_spark = df_spark.withColumn("age", year("trans_date_trans_time") - year("dob"))

# Select cleaned and relevant columns
final_cols = ['cc_num', 'merchant', 'category', 'amt', 'gender', 'city', 'state', 'lat', 'long',
              'city_pop', 'job', 'age', 'trans_hour', 'trans_dayofweek', 'merch_lat', 'merch_long', 'is_fraud']
df_spark = df_spark.select(*final_cols)

# Preview
df_spark.show(5)
```

```python
from pyspark.sql.functions import col, rand

# Separate fraud and non-fraud transactions
fraud_df = df_spark.filter(col("is_fraud") == 1)
nonfraud_df = df_spark.filter(col("is_fraud") == 0)

# Take exactly 50,000 non-fraud rows (shuffled)
nonfraud_sampled_df = nonfraud_df.orderBy(rand()).limit(50000)

# Combine and shuffle the final balanced dataset
balanced_df = fraud_df.union(nonfraud_sampled_df).orderBy(rand())

# Show class counts to confirm
balanced_df.groupBy("is_fraud").count().show()

```

# Preparation for Data Modeling

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline

# Step 1: Define categorical columns
categorical_cols = ["gender", "category", "job", "state"]

# Step 2: Index and encode categoricals
indexers = [StringIndexer(inputCol=col, outputCol=col+"_index", handleInvalid="keep") for col in categorical_cols]
encoders = [OneHotEncoder(inputCol=col+"_index", outputCol=col+"_encoded") for col in categorical_cols]

# Step 3: Define numeric columns to include
numeric_cols = ["amt", "city_pop", "age", "trans_hour", "trans_dayofweek", "lat", "long", "merch_lat", "merch_long"]

# Step 4: Combine all features
assembler_inputs = [col+"_encoded" for col in categorical_cols] + numeric_cols
assembler = VectorAssembler(inputCols=assembler_inputs, outputCol="features")

# Step 5: Build pipeline
pipeline = Pipeline(stages=indexers + encoders + [assembler])

# Step 6: Fit and transform the data
model_data = pipeline.fit(balanced_df).transform(balanced_df)

# Step 7: Split into train/test
train_data, test_data = model_data.randomSplit([0.8, 0.2], seed=42)

# # Check counts
print(f"Train count: {train_data.count()}, Test count: {test_data.count()}")
```

```python
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Common evaluator
evaluator = BinaryClassificationEvaluator(labelCol="is_fraud", metricName="areaUnderROC")

# -----------------------------
# 1. Logistic Regression
# -----------------------------
lr = LogisticRegression(featuresCol="features", labelCol="is_fraud")
lr_param_grid = ParamGridBuilder()\
    .addGrid(lr.regParam, [0.01, 0.1, 1.0])\
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])\
    .build()
lr_cv = CrossValidator(estimator=lr,
                       estimatorParamMaps=lr_param_grid,
                       evaluator=evaluator,
                       numFolds=3,
                       parallelism=2)
lr_model = lr_cv.fit(train_data)
lr_predictions = lr_model.transform(test_data)
lr_auc = evaluator.evaluate(lr_predictions)

# -----------------------------
# 2. Random Forest
# -----------------------------
rf = RandomForestClassifier(featuresCol="features", labelCol="is_fraud", seed=42)
rf_param_grid = ParamGridBuilder()\
    .addGrid(rf.numTrees, [20, 50])\
    .addGrid(rf.maxDepth, [5, 10])\
    .build()
rf_cv = CrossValidator(estimator=rf,
                       estimatorParamMaps=rf_param_grid,
                       evaluator=evaluator,
                       numFolds=3,
                       parallelism=2)
rf_model = rf_cv.fit(train_data)
rf_predictions = rf_model.transform(test_data)
rf_auc = evaluator.evaluate(rf_predictions)

# -----------------------------
# 3. Gradient Boosted Trees
# -----------------------------
gbt = GBTClassifier(featuresCol="features", labelCol="is_fraud", seed=42)
gbt_param_grid = ParamGridBuilder()\
    .addGrid(gbt.maxDepth, [5, 7])\
    .addGrid(gbt.maxIter, [10, 20])\
    .build()
gbt_cv = CrossValidator(estimator=gbt,
                        estimatorParamMaps=gbt_param_grid,
                        evaluator=evaluator,
                        numFolds=3,
                        parallelism=2)
gbt_model = gbt_cv.fit(train_data)
gbt_predictions = gbt_model.transform(test_data)
gbt_auc = evaluator.evaluate(gbt_predictions)

# -----------------------------
# Print Results
# -----------------------------
print(f"\n Logistic Regression AUC: {lr_auc:.4f}")
print(f"   Best Params: regParam={lr_model.bestModel._java_obj.getRegParam()}, "
      f"elasticNetParam={lr_model.bestModel._java_obj.getElasticNetParam()}")

print(f"\n Random Forest AUC: {rf_auc:.4f}")
print(f"   Best Params: numTrees={rf_model.bestModel.getNumTrees}, "
      f"maxDepth={rf_model.bestModel.getOrDefault('maxDepth')}")

print(f"\n Gradient Boosted Trees AUC: {gbt_auc:.4f}")
print(f"   Best Params: maxDepth={gbt_model.bestModel.getOrDefault('maxDepth')}, "
      f"maxIter={gbt_model.bestModel.getOrDefault('maxIter')}")
```
